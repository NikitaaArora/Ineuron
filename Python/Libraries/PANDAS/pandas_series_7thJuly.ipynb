# Merging and joining dataframes

# Concatenation

```python
df1 = pd.DataFrame({'A' : ['A0','A1']  ,

                    'B' : ['B0','B1']})

df2 = pd.DataFrame({  'A': ['A2','A3'],
                      'B' : ['B2','B3']
})
                

```

```python
df1
```

```python
df2
```

```python
result = pd.concat([df1,df2])
result
```

```python
result1 = pd.concat([df2,df1])
result1
```

# Merging

```python
left = pd.DataFrame({'key': ['K0','K1'],
                     'A': ['A0','A1']})

right = pd.DataFrame({
    'key':['K0','k2'],
    'b' : ['b0','b2']
})
```

```python
left
```

```python
right
```

```python
result = pd.merge(left,right, on= "key",how = 'inner')#intersect
result
```

```python
#Joining
left = pd.DataFrame({'A':['b0','A1']}, index = ['K0','K1'])
left
```

```python
right = pd.DataFrame({'B':['b0','b1']}, index = ['K0','k2'])
right
```

```python
left.join(right,how='right')
```
