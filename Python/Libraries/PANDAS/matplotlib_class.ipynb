```python
#plotting the Age distribution
plt.plot(titanic['Age']) #line plot
plt.title('Age Distibution')
plt.xlabel('Passenger_Index')
plt.ylabel('Age')
plt.show()
```

# Plotting with Different Styles


```python
#this is a horrible graph as it is not telling anything.

#customizing the plot
plt.plot(titanic['Age'],linestyle = '--',marker = 'o',color = 'r')
plt.title('Customised Age Distribution')
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.show()
```

```python
#Customizing plots
#Adding grid lines, adjusting plot limits and adding legends. 

plt.plot(titanic['Age'],color = 'g')
plt.title('Age distribution with Grid')
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.grid(True)
plt.xlim(0,100)
plt.ylim(0,80)
plt.legend(['Age'])
plt.show()
```

```python
#creating multiple subplots
fig,axs = plt.subplots(2,2)
axs[0,0].plot(titanic['Age'],color = 'b')
axs[0,0].set_title('Age Distribution')
axs[0,1].plot(titanic['Fare'],color = 'r')
axs[0,1].set_title('Fare Distribution')
axs[1,0].plot(titanic['SibSp'],color = 'g')
axs[1,0].set_title('SibSp Distribution')
axs[1,1].plot(titanic['Parch'],color='m')
axs[1,1].set_title('Parents/Childer Abroad')

fig.tight_layout()
plt.show()

```

# for example there is a series of numbers:

data: 4,5,9,10,11,13,14,19

Ways to find bins:

1st Method:

Maximum is 19,
minimum is 4
ie. bin-width = (19-4)/total_number_of_bins_you_want = (19-4)/3 = (15)/3 = 5  
bin1 : 4+5 = 9 ie. [4,9)
bin2 : 9+5 =14 ie. [9,14)
bin3 : 14+5= 19 ie. [14,19]
Next, sort the data into these bins:
Bin 1: 4, 5
Bin 2: 9, 10, 11, 13
Bin 3: 14, 19

2nd Method: 
on the basis of the frequency

ie. I want three bins. so that would be

bin1: 4,5,9
bin2: 10,11,13
bin3: 14,19

```python
import matplotlib.pyplot as plt

# Data and bins
data = [4, 5, 9, 10, 11, 13, 14, 19]
bins = [4, 9, 14, 19]

# Plot histogram
plt.hist(data, bins=bins, edgecolor='black', alpha=0.7)

# Adding titles and labels
plt.title("Histogram of Data with Bin Width 5")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# Show plot
plt.show()
```
