```python
#fucntion that applies sir/madam according to the gender

def append_name(row): #gives a dataframe
    if row['Survived'] == 1:
        return "Mr " + row['Name']
    
    else:

        return "Ms " + row["Name"]
```

```python
titanic.apply(append_name,axis = 1) # axis = 1 is across rows .

# this essentially makes a new column so will use axis =1 only. 

```

```python
titanic['Chnaged_Names'] = titanic.apply(append_name,axis=1)
titanic['Chnaged_Names']
```
