```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [10,5,2,6,3]
```

```python
fig = plt.figure()  

ax = fig.add_axes([0,0,2,1])

ax.plot(x,y)
ax.set_title('X vs Y')
ax.set_xlim(-4,4)
ax.set_xlabel('X values')
ax.set_label('X vs Y')
```

```python
#subplots
fig,ax = plt.subplots(3,2) #firgure just creates the blank canvas rest everything we do using ax.

ax[0,0].scatter(x,y)

ax[0,1].bar(x,y)

ax[1,0].plot(x,y)
```
